```python

path = "../../data/uncinatus_fdrgb.nii" # (121, 145, 121, 3) shows the axons in the uncinate fasciculus. the last dimension is the RGB values (?)
path = "../../data/exCapsule.nii" # (121, 145, 121) shows the axons in the external capsule. 
path = "../../data/ifof_fdrgb.nii" # (121, 145, 121, 3) shows the axons in the inferior fronto-occipital fasciculus. the last dimension is the RGB values (?)


path = "../../data/MNI_T1_lowres.nii" #(91, 109, 91, 1). floats in (0,255) corresponding to shade
path = "../../data/atlas116.nii" #(91, 109, 91). int in [0,116] corresponding to zones

```

```python
img = nib.load(path).get_fdata()
print(img.shape)
print(f"The .nii files are stored in memory as numpy's: {type(img)}.")
```

```python
plt.style.use('default')
fig, axes = plt.subplots(4,4, figsize=(12,12))
for i, ax in enumerate(axes.reshape(-1)):
    ax.imshow(img[:,:,5*i])
plt.show()
```
