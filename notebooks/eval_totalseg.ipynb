```python
%load_ext autoreload
%autoreload 2

from src.config import config
from pathlib import Path
import nibabel as nib
from src.totalseg_dataloader import get_dataloader

data_dir = Path(config["DATA_DIR"]) / "TotalSeg" 
```

```python
# get all label_ids
# list all files like data_dir/*/segmentations/*.nii.gz
label_ids = []
for organ_dir in (data_dir).iterdir():
    if organ_dir.is_dir():
        segmentations_dir = organ_dir / "segmentations"
        for seg_file in segmentations_dir.glob("*.nii.gz"):
            label_id = seg_file.name.split(".")[0]
            if label_id not in label_ids:
                label_ids.append(label_id)
```

```python
# save dict to ic_segmentation/data/label_ids_totalseg.py
with open("ic_segmentation/data/label_ids_totalseg.py", "w") as f:
    f.write("label_ids = [\n")
    for label_id in label_ids:
        f.write(f'    "{label_id}",\n')
    f.write("]\n")
```

```python
stats_dict_path = Path(config["DATA_DIR"]) / "TotalSeg" /"dataset_stats.json"
# open json file
import json
with open(stats_dict_path, 'r') as f:
    stats_dict = json.load(f)

empty_segmentations = stats_dict['empty_segmentations']
print(empty_segmentations)
```

```python
import pandas as pd
run1 = api.run("/tidiane/ic_segmentation/runs/iwrk20z9")
run2 = api.run("/tidiane/ic_segmentation/runs/wv1t7mpm")
df = pd.DataFrame([run1.config, run2.config]).transpose()

df.columns = [run1.name, run2.name]
print(df[df[run1.name] != df[run2.name]])

run1_metrics = run1.history()
run2_metrics = run2.history()

conf_compared = "sampling_method"
```

```python
# dice comparison plot per label
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Combine the dataframes with a run identifier
run1_metrics_copy = run1_metrics.copy()
run1_metrics_copy['run'] = run1.config[conf_compared]  # Set your custom name here
run2_metrics_copy = run2_metrics.copy()
run2_metrics_copy['run'] = run2.config[conf_compared]  # Set your custom name here

combined_metrics = pd.concat([run1_metrics_copy, run2_metrics_copy])

plt.figure(figsize=(10, 6))
sns.boxplot(x="label_id", y="dice", hue="run", data=combined_metrics, palette=['lightblue', 'lightgreen'])
plt.title("Dice Score Distribution per label_id")
plt.xlabel("Label ID")
plt.xticks(rotation=45)
plt.ylabel("Dice Score")
plt.legend(title="Run")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
combined_metrics["avg_context_to_target_dice_scores"] = combined_metrics["context_to_target_dice_scores"].apply(lambda x: sum(x)/len(x) if isinstance(x, list) and len(x) > 0 else None)
sns.scatterplot(x="avg_context_to_target_dice_scores", y="dice", hue="run", data=combined_metrics)
plt.title("Dice Score vs Avg Dice Context to Target Dice Scores")
plt.xlabel("Avg Dice Context to Target Dice Scores")
plt.ylabel("Dice Score")
plt.legend(title="Run")
plt.show()
```

```python
# same data but bin x axis
plt.figure(figsize=(10, 6))
sns.barplot(x=pd.cut(combined_metrics["avg_context_to_target_dice_scores"], bins=10), y="dice", hue="run", data=combined_metrics, palette=['lightblue', 'lightgreen'])
plt.title("Dice vs Context-Target GT Dice")
plt.xlabel("Context-Target GT Dice")
plt.ylabel("Dice Score")
plt.xticks(rotation=45)
plt.legend(title="Run")
plt.show()
```
