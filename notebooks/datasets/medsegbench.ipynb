```python
import numpy as np
ds_path = "/nfs/data/nii/data1/Analysis/camaret___in_context_segmentation/ANALYSIS_20251122/data/MedSegBench/abdomenus_512.npz"
data = np.load(ds_path)
```

```python
print(data["train_images"].shape)
print(data["train_label"].shape)
```

```python
labels_per_case = []
for train_mask in data["train_label"]:
    labels_per_case.append(np.unique(train_mask))
```

```python
# average number of labels per case
avg_labels = np.mean([len(labels) for labels in labels_per_case])
print(f"Average number of labels per case: {avg_labels}")
```
